```python
import datetime

#回傳今天的日期：yyyymmdd
def Today():
    date = str(datetime.date.today()).replace("-","")
    return date

#回傳迭代的日期
def DateItertion(target_date):
    day_del = datetime.timedelta(days=1)
    target_date = datetime.datetime.strptime(target_date,"%Y%m%d")
    d = str(target_date - day_del).replace("-","")
    return d[0:8]
```

```python
#today = str(Today())
OUTPUT_FILE = '/Users/dayuan/Desktop/爬蟲/20230306_udn.json'
OUTPUT_FILE_C = '/Users/dayuan/Desktop/爬蟲/NewsDaliyCount.csv'
target_date = '20180731'

while target_date !='20180630':
    driver = webdriver.Safari()
    link = "https://udn.com/news/archive?date="+target_date
    driver.get(link)
    time.sleep(4)

    results = []
    for scroll in range(60):   
        #if scroll%20 == 1 :print(scroll,datetime.datetime.today())
        try:
            driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
            time.sleep(randint(5,7))   
        except:
            print("break")
            break 

    elem = driver.find_elements(By.CLASS_NAME,"story-list__news")
    e_count = 0
    for e in elem:
        try:
            result={}
            url = e.find_element(By.TAG_NAME,"a").get_attribute('href')
            if e_count <20:
                title = e.find_element(By.TAG_NAME,"a").get_attribute('aria-label')
                y = target_date[:4]
                date = y+'-'+e.find_element(By.CLASS_NAME,"story-list__time").text
            else:
                title = e.find_element(By.TAG_NAME,"a").get_attribute('title')
                date = e.find_element(By.CLASS_NAME,"story-list__time").text

            domain = e.find_element(By.CLASS_NAME,"story-list__cate").text
            view = e.find_element(By.CLASS_NAME,"story-list__view").text

            result = {
                'date':date,
                'domain':domain,
                'title':title,
                'view':view,
                'url':url
            }        
            results.append(result)
            e_count+=1

        except:
            print("Error")
            continue
            
    with open(OUTPUT_FILE_C,'a', encoding='utf-8') as ofc:
        writer = csv.writer(ofc)
        writer.writerow([target_date,len(results)])
        
    with open(OUTPUT_FILE,'a', encoding='utf-8') as ofile:
        try:
            print(len(results))
            json.dump(results, ofile, indent=2,sort_keys=True, ensure_ascii=False)
            print(target_date,datetime.datetime.today())
            target_date = DateItertion(target_date) #日期改成前一天
        except:
            print("File Error")
    
    driver.quit()
    time.sleep(randint(50,100)) #要爬下一天時，隨機等待
```
